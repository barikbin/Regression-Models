# Spliting the dataset

```python
# First we find highly corrolated features

corr = train.corr()
high_corr_features = corr.index[abs(corr['SalePrice']) > 0.52]
print(f'Highly Corrolated Features:\n', high_corr_features, '\n')
print(f'No. of Highly Corrolated Features:' , len(high_corr_features))
```

```python

X = train[high_corr_features.drop('SalePrice')]
y = train[['SalePrice']]

print(X)
print(y)
```

```python
# Here I should add explnations

test = test[high_corr_features.drop('SalePrice')]
```

```python

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Printing the X_train Features shape
print(f'X_train', X_train.shape)

# Printing the y_train Labels shape
print(f'y_train', y_train.shape)

# Printing the x_test Features shape
print(f'x_val', X_val.shape)

# Printing the y_test Label shape
print(f'y_val', y_val.shape)
```

```python

test.fillna(test.mode().iloc[0], inplace=True)
```

```python

# Normalize the data
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_val = ss.transform(X_val)
test = ss.transform(test)
```

# The random Forest Model

```python
# Building the Model
# Here we create the Random Fforest Regressor Model
clf_rfr = RandomForestRegressor(random_state=0)

#Training the Model with X_train and y_train
clf_rfr.fit(X_train, y_train)

# Predicting on the X_test
pred = clf_rfr.predict(X_val)


# Testing the moel accuracy with different paramters

# Getting te R_squared
r2 = r2_score(y_val, pred)

# Getting the Mean Absolute Error
mae = mean_absolute_error(y_val, pred)

# Getting the Mean Squared Error
mse = mean_squared_error(y_val, pred)

# Getting the Root Mean Squared Error
rmse = np.sqrt(mse)

# Putting the evluation into the dataframe
rfr_results = pd.DataFrame([['RandomForestRegressor', r2, mae, mse, rmse]], columns=['Model', 'R-squared', 'Mean Absolute Error', 'Mean Squared Error', 'Root Mean Squared Error'])

result = rfr_results.append(rfr_results, ignore_index=True)

result  # Here we print the model with different Metrics
```
